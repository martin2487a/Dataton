```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar los datasets
postulantes = pd.read_csv("Datos_abiertos_admision_2021_1_2024_1.csv")
```

```python
postulantes.head()
```

```python
# Tasa de adminisión por colegio

admisiones = postulantes[postulantes["INGRESO"] == "SI"]
admisiones
```

```python
admisiones_por_colegio_por_año = admisiones["ANIO_POSTULA"].value_counts()
admisiones_por_colegio_por_año.sort_index(ascending=True)
```

```python
admisiones_por_colegio_por_año.plot(kind="bar")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

admitidos_por_colegio = pd.DataFrame(admisiones.groupby("COLEGIO")["INGRESO"].size().sort_values(ascending=False))

admitidos_por_colegio
```

```python
from matplotlib import pyplot as plt
admitidos_por_colegio['INGRESO'].plot(kind='line', figsize=(16, 4), title='INGRESO')
plt.gca().spines[['top', 'right']].set_visible(False)
plt.xticks(rotation=45)
plt.show()
```

```python
# En la presente gráfica, se observa que de todos los colegios del Perú el Colegio Trilce de los Olivos ubicado
# en el departamento y provincia de Lima, es el que cuenta con el mayor número de estudiantes,
# ingresantes a la Universidad Nacional de Ingeniería con 124 estudiantes desde el periodo 2021 al 2024
```
